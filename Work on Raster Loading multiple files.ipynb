```r
library(raster)

# set working directory to the directory containing TIFF files
setwd("100fps_80px/")

# create a list of all TIFF files in the directory
tif_files <- list.files(pattern = "\\.tif$")

# use stack to load all TIFF files into a single raster stack
raster_stack <- stack(tif_files)
```

```r
library(hsdar)

# convert raster stack to a hyperspectral object
hyperspec <- rasterToSpectra(raster_stack)

# print information about the hyperspectral object
print(hyperspec)

```
